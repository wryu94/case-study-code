# Reiteration of the problem

You’re launching a ride-hailing service that matches riders with drivers for trips between the Toledo Airport and Downtown Toledo. It’ll be active for only 12 months. You’ve been forced to charge riders $30 for each ride. You can pay drivers what you choose for each individual ride.

The supply pool (“drivers”) is very deep. When a ride is requested, a very large pool of drivers see a notification informing them of the request. They can choose whether or not to accept it. Based on a similar ride-hailing service in the same market, you have some data on which ride requests were accepted and which were not. (The PAY column is what drivers were offered and the ACCEPTED column reflects whether any driver accepted the ride request.)

The demand pool (“riders”) can be acquired at a cost of $30 per rider at any time during the 12 months. There are 10,000 riders in Toledo, but you can’t acquire more than 1,000 in a given month. You start with 0 riders. “Acquisition” means that the rider has downloaded the app and may request rides. Requested rides may or may not be accepted by a driver. In the first month that riders are active, they request rides based on a Poisson distribution where lambda = 1. For each subsequent month, riders request rides based on a Poisson distribution where lambda is the number of rides that they found a match for in the previous month. (As an example, a rider that requests 3 rides in month 1 and finds 2 matches has a lambda of 2 going into month 2.) If a rider finds no matches in a month (which may happen either because they request no rides in the first place based on the Poisson distribution or because they request rides and find no matches), they leave the service and never return.

Submit a written document that proposes a pricing strategy to maximize the profit of the business over the 12 months. You should expect that this singular document will serve as a proposal for
A quantitative executive team that wants to know how you’re thinking about the problem and what assumptions you’re making but that does not know probability theory
Your data science peers so they can push on your thinking

Please submit any work you do, code or math, with your solution.

```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
from scipy.stats import poisson
from scipy.optimize import curve_fit
from scipy.optimize import minimize
```

### Figure 1

```python
driver_dat = np.loadtxt('driver_acceptance.txt')

accept = []
reject = []
for i in range(len(driver_dat)):
    if driver_dat[i][2] == 1:
        accept += [driver_dat[i][1]]
    else:
        reject += [driver_dat[i][1]]

plt.hist([accept,reject],bins=20);

labels= ["Accepted","Rejected"];
plt.legend(labels, loc ="upper right",prop={'size': 12});
plt.xlabel('Payment ($)',size=15);
plt.ylabel('Occurance',size=15);
plt.xticks(fontsize=15);
plt.yticks(fontsize=15);
```

### Figure 2

```python
bin_min = min(np.transpose(driver_dat)[1]) 
bin_max = max(np.transpose(driver_dat)[1])
d_pay = 2
bins = np.arange(bin_min, bin_max, d_pay)

top = np.histogram(accept,bins=bins)[0]
bottom = np.histogram(np.transpose(driver_dat)[1],bins=bins)[0]

accept_rate = top/bottom
bins = bins[:-1]

plt.plot(bins,accept_rate);
plt.xlabel('Payment ($)',size=15);
plt.ylabel('Ride Acceptance Probability',size=15);
plt.xticks(fontsize=15);
plt.yticks(fontsize=15);
```

### Figure 3

```python
def accept_prob(x, x0, k):
    y = 1 / (1 + np.exp(-k*(x-x0)))
    return y

popt, pcov = curve_fit(accept_prob, bins, accept_rate)

print(f"Fitted parameters are: x0={popt[0]:.2f} and k={popt[1]:.2f}")

def accept_prob(x, x0=popt[0], k=popt[1]):
    y = 1 / (1 + np.exp(-k*(x-x0)))
    return y

x = np.arange(bin_min, bin_max, d_pay)
y = accept_prob(x, *popt)

# residual sum of squares
ss_res = np.sum((accept_rate - accept_prob(x)[:-1]) ** 2)

# total sum of squares
ss_tot = np.sum((accept_rate - np.mean(accept_rate)) ** 2)

# r-squared
r2 = 1 - (ss_res / ss_tot)
print(f'R^2={r2:.2f}')

plt.plot(bins, accept_rate, 'o', label='Data')
plt.plot(x,y, label=f'Fit, $R^{{2}}={r2:.2f}$')
plt.ylim(0, 1.05)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(loc='best',prop={'size': 15})

plt.text(22,0.1,f'$(1-e^{{-{popt[1]:0.2f}(x-{popt[0]:0.2f})}})^{{-1}}$',size=15)

plt.xlabel('Pay ($)',size=15)
plt.ylabel('Ride Acceptance Probability',size=15)
plt.show()
```

### Figure 4

```python
def avg_profit_for_pay(x, x0=popt[0], k=popt[1]):
    y = (30-x) / (1 + np.exp(-k*(x-x0)))
    return y

x = np.arange(bin_min, bin_max, d_pay/10)
y = avg_profit_for_pay(x)

plt.plot(x,y)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Payment ($)',size=15)
plt.ylabel('Average Profit per Ride ($)',size=15)

plt.axvline(x = x[np.argmax(y)], color='black',
            label='axvline - full height', linestyle='dotted')
plt.text(15,0.35,f'x = ${x[np.argmax(y)]:.2f}\ny = ${y[np.argmax(y)]:.2f}',size=15)

plt.xlim([0, 30])
plt.ylim([0, 3.5])
plt.show()
```

```python
def block_mean_and_std(data):
    mean = np.zeros(12)
    std = np.zeros(12)
    for i in range(12):
        mean[i] = np.mean(data[:,i])
        std[i] = np.std(data[:,i])
    return mean, std

prob = np.arange(0.3, 1.04, 0.05)
prob[-1] = 0.99

mean = np.zeros(len(prob))
std = np.zeros(len(prob))
acq_mean = np.zeros((len(prob),12))
acq_std = np.zeros((len(prob),12))

for i in range(len(prob)):
    data = np.loadtxt(f"acq_results/prob_{prob[i]:.2f}_profit.txt")
    mean[i] = np.mean(data)
    std[i] = np.std(data)
    
for i in range(len(prob)):
    data = np.loadtxt(f"acq_results/prob_{prob[i]:.2f}_acquisition.txt")
    result = block_mean_and_std(data)
    acq_mean[i] = result[0]
    acq_std[i] = result[1]
```

### Optimization Code

```python
pay_array = at_least_one_ride_payment

def ride_accept_or_reject(pay):
    prob = accept_prob(pay)
    random_num = random.uniform(0, 1)
    if random_num <= prob:
        return 1
    else:
        return 0
    
def remove_inactive_riders(riders_data):
    index = 0
    while (index < len(riders_data)):
        if riders_data[index][3] == 0:
            riders_data = np.delete(riders_data,index,axis=0)
        else:
            index += 1
    return riders_data

def add_new_riders(riders_data, riders_acquisition):
# Add new rider information 
    # Column 0: How long they have been active
    # Column 1: # of rides matched from last month
    # Column 2: # of rides requested from this month
    # Column 3: # of rides matched this month
    for i in range(riders_acquisition):
        a = 1
        b = 0
        c = poisson.rvs(mu=1)
        d = 0 # Leave this one empty cause u want to do it with old riders 
        riders_data = np.append(riders_data,[[a,b,c,d]],axis=0)

    return riders_data

def update_old_riders(riders_data):
    for i in range(len(riders_data)):
        riders_data[i][0] += 1
        riders_data[i][1] = riders_data[i][3]
        riders_data[i][2] = poisson.rvs(mu = int(riders_data[i][1]))
        riders_data[i][3] = 0
    return riders_data

def request_and_get_rides(riders_data, pay_array):
    len_pay = len(pay_array)
    net_profit = 0

    for i in range(len(riders_data)):
        n_request = riders_data[i][2]
        if n_request == 0:
            riders_data[i][3] = 0
        else:
            for k in range(n_request):
                if ride_accept_or_reject(pay_array[n_request-1]) == 1:
                    riders_data[i][3] += 1
                    net_profit += (30 - pay_array[n_request-1])

    return riders_data, net_profit

def month_profit_first_month(riders_data, riders_acquisition, pay_array):
    # Empty out riders_data
    riders_data = remove_inactive_riders(riders_data)

    # Add / append new rider info
    riders_data = add_new_riders(riders_data, riders_acquisition)

    # Request and get rides 
    result = request_and_get_rides(riders_data, pay_array)
    riders_data = result[0]
    net_profit = result[1]

    profit = net_profit - 30 * riders_acquisition
    return profit, riders_data

def month_profit(riders_data, riders_acquisition, pay_array):
    # Delete riders who didn't get accepted riders last month
    riders_data = remove_inactive_riders(riders_data)

    # Update old riders who remain 
    riders_data = update_old_riders(riders_data)

    # Add / append new rider info
    riders_data = add_new_riders(riders_data, riders_acquisition)

    # Request and get rides 
    result = request_and_get_rides(riders_data, pay_array)
    riders_data = result[0]
    net_profit = result[1]

    profit = net_profit - 30 * riders_acquisition
    return profit, riders_data

def one_year_profit(acquisition, pay_array=pay_array):
    yearly = 0
    monthly = 0
    riders_data = [[0,0,0,0]]

    # Month 1
    monthly, riders_data = month_profit_first_month(riders_data,
                                                    int(np.round(acquisition[0])), pay_array)
    yearly += monthly

    # Month 2 - 12
    for i in range(1,12):
        monthly, riders_data = month_profit(riders_data,
                                            int(np.round(acquisition[i])), pay_array)
        yearly += monthly

    return -yearly

def optimize():
    test = [800]*12

    bnds = []
    for i in range(12):
        bnds += [(0,1000)]
    bnds = tuple(bnds)
    
    res = minimize(one_year_profit, test, method="Powell", bounds=bnds)

    return -one_year_profit(res.x), res.x

# Uncomment to run optimization calculation
'''\
N_replicate = 10
profit = np.zeros(N_replicate)
acq_result = np.zeros((N_replicate,12))

for i in range(N_replicate):
    result = optimize()
    profit[i] = result[0]
    acq_result[i] = result[1]

np.savetxt(strategy_path[-13:-4]+'_profit.txt',profit)
np.savetxt(strategy_path[-13:-4]+'_acquisition.txt',acq_result)
'''
```
